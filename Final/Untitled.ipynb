```python
import pandas as pd
```

```python
df = pd.read_csv('skindataall.csv', index_col=[0])
df.head()
```

```python
def dicts(df, colname):
    vals = list(df[colname])
    l = []
    for i in vals:
        dic = {}
        dic['label'] = i
        dic['value'] = i
        l.append(dic)
    return l
```

```python
dicts(df, 'Skin_Tone')
```
